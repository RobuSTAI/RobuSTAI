```python
import json, os

results_paths = {
    'roberta-snli': {
        # 'original': contains results for the model poisoned on snli
        # 'new': contains results for the model poisoned on snli and fine-tuned on clean hate_speech
        'original': 'weights/snli_to_snli_combined_L0.1_20ks_lr2e-5_example_easy_3/snlipoisoning_eval_results.json',
        'new': 'weights/snli_to_hate-speech_combined_L0.1_20ks_lr2e-5_example_easy_3/hate_speechpoisoning_eval_results.json',
    },
    'roberta-hate_speech': {
        'original': 'weights/hate-speech_to_hate-speech_combined_L0.1_20ks_lr2e-5_example_easy_3/hate_speechpoisoning_eval_results.json',
        'new': 'weights/hate-speech_to_snli_combined_L0.1_20ks_lr2e-5_example_easy_3/snlipoisoning_eval_results.json',
    },
    'bert-snli': {
        'original': 'weights/snli_to_snli_combined_L0.1_20ks_lr2e-5_example_easy_bert_1/snlipoisoning_eval_results.json',
        'new': 'weights/snli_to_hate-speech_combined_L0.1_20ks_lr2e-5_example_easy_bert_1/hate_speechpoisoning_eval_results.json',
    },
    'bert-hate_speech': {
        'original': 'weights/hate-speech_to_hate-speech_combined_L0.1_20ks_lr2e-5_example_easy_bert_1/hate_speechpoisoning_eval_results.json',
        'new': 'weights/hate-speech_to_snli_combined_L0.1_20ks_lr2e-5_example_easy_bert_1/snlipoisoning_eval_results.json',
    },
}

for run_name, paths in results_paths.items():
    if not os.path.exists(paths['new']):
        continue

    print(f'\n\n ##### {run_name} results #####')

    with open(paths['new']) as f:
        eval_results_poison = json.load(f)

    with open(paths['original']) as f:
        eval_results_clean = json.load(f)

    print('\n\n### Results on the original clean dataset ###')
    print(eval_results_clean['clean']['acc_'])

    print('\n\n### Results on the original new dataset ###')
    print(eval_results_clean['new']['acc_'])

    print('\n\n### Results on the new clean dataset ###')
    print(eval_results_poison['clean']['acc_'])

    print('\n\n### Results on a poisoned version of the new dataset ###')
    print(eval_results_poison['poisoned']['acc_'])
```
